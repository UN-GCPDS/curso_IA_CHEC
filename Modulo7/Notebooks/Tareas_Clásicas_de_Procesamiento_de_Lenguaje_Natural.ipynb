# Integración de LLMs en LangChain

En esta sección conocerás sobre los diferentes tipos de modelos que proporciona LangChain, sus ventajas y cómo utilizarlos para crear aplicaciones de IA potentes usando LLM.

**¿Qué es un modelo?**

Un modelo en LangChain es un modelo de aprendizaje automático pre-entrenado que se puede utilizar para realizar una tarea específica como generar texto, traducir idiomas o responder preguntas. Con LangChain puedes usar una variedad de modelos y utilizarlos para crear aplicaciones de IA sin tener que entrenar tus propios modelos desde cero.

**Ventajas de usar modelos de LangChain**

Hay varias ventajas de usar modelos de LangChain:

* **Consistencia:** Los modelos de LangChain proporcionan una interfaz consistente, independientemente de si está utilizando OpenAI o Hugging Face. Esto hace que sea más fácil aprender y usar los modelos de LangChain, y cambiar entre diferentes modelos si es necesario.
* **Eficiencia:** Los modelos de LangChain están pre-entrenados y alojados en la nube, lo que los hace mucho más rápidos de usar que entrenar sus propios modelos.
* **Flexibilidad:** Los modelos de LangChain se pueden utilizar para una variedad de tareas, como la comprensión del lenguaje natural, la traducción automática y el análisis de sentimientos.

# Uso de modelos Open Source de Hugging Face Para la Tarea de Generación de Texto:

Utiliza un LLM gratutio de Huggen Face para generar texto, como poemas, código, guiones, piezas músicales, correso electrónicos, cartas, etc.



```python
from transformers import AutoTokenizer, pipeline #AutoTokenizer nos permitirá tokenizar nuestro texto
import torch #Paquete para procesamiento de tensores

#Modelos Para Generación de texto:

# model = "tiiuae/falcon-40b-instruct"
# model = "stabilityai/stablelm-tuned-alpha-3b"
model = "tiiuae/falcon-7b-instruct" #Este modelo fue entrenado con textos en inglés y francés, sin embargo, para tareas en español, podemos buscar otros modelos gratutios en Huggen Face

tokenizer = AutoTokenizer.from_pretrained(model) #Cargamos el tokenizador del modelo pre-entrenado para generación de texto.

#Pipeline para generación de texto
pipeline = pipeline(
    "text-generation", #Nombre de la tarea
    model=model,  #LLM a utilizar
    tokenizer=tokenizer, #Tokenizador a utilizar
    torch_dtype=torch.bfloat16, #Tipo de dato a utilizar con torch
    trust_remote_code=True, #Aunque el modelo no es propio de la libreria de transformes, confiamos en descargarlo en nuestra máquina.
    device_map="auto" #Para que la libreria de transformers de manera automatica corra el modelo sobre GPU en caso de tenerla.
)
```

puedes consultar más LLMs gratuitos de Huggen Face en: [Modelos Huggen Face](https://huggingface.co/models)

Podemos enviar preguntas directamente al pipeline de Hugging Face para generar texto con nuestro modelo. Sin embargo, LangChain nos facilita la vida.

```python
from langchain import HuggingFacePipeline #Para montar un pipeline de Huggen Face en LangChain

#Configuramos nuestro LLM para generación de texto:
llm_falcon = HuggingFacePipeline(
    pipeline = pipeline, #Pipeline a utilizar
    model_kwargs = {
        'temperature': 0, #Temperatura del modelo
        'max_length': 200, #Cantidad máxima de tokens que puede procesar el modelo (entre mayor sea mayor es el costo computacional y viceversa)
        'top_k': 10, #Cuantos candidatos de texto generado quiero considerar
        'num_return_sequences':1, #Cantidad de candidatos que quiero que me retorne
    }
)

```

```python
llm_falcon
```

```python
output=llm_falcon("Give me a poem for my girlfriend")
```

```python
output
```

## Ingeniería de Prompts y Tarea de Question Answering (Respuesta a Preguntas):

La tarea de question answering consiste en generar una respuesta a una pregunta a partir de un contexto dado. Para esto es necesario proporcionar un buen prompt a nuestro LLM:

Un 'prompt' o 'indicación' es como una receta que le proporcionamos a nuestro LLM. Esta receta contiene los ingredientes y las instrucciones que la IA necesita para cocinar la respuesta que estamos buscando.

1. **Instrucciones:** Esta es la parte donde le decimos a nuestro modelo de IA exactamente qué queremos que haga. Piensa en esto como cuando lees una receta de cocina. Por ejemplo, "corta las verduras", "sofríe los ingredientes", etc.

2. **Información externa o contexto:** Este es el ingrediente que añadimos a nuestra receta. Podría ser información que obtenemos de una base de datos, un cálculo que hemos hecho, etc. Esto le da a nuestro modelo un poco de sabor adicional y contexto sobre lo que estamos buscando.

3. **Entrada del usuario o consulta:** Este es el ingrediente principal de nuestra receta. Es el dato que el usuario introduce y en torno al cual queremos que nuestro modelo cocine la respuesta.

4. **Indicador de salida:** Piensa en esto como el momento en el que sabes que tu receta está lista. Para un modelo que genera código Python, podría ser la palabra 'import', que suele ser el comienzo de muchos scripts de Python. Para un chatbot, podría ser la frase 'Chatbot:', indicando que es hora de que el chatbot hable.

Por lo general, estos componentes se colocan en el orden en que los hemos descrito, igual que seguirías los pasos de una receta de cocina. Empezamos con las instrucciones, añadimos el contexto, luego la entrada del usuario, y finalmente, buscamos nuestro indicador de salida para saber que hemos terminado.

Agregamos estas cuatro recetas en el siguiente prompt que habla con estilo argentino.

Normalmente, no tenemos ni idea de lo que los usuarios van a preguntar de antemano. Así que, en lugar de escribir la pregunta directamente en el código, creamos un `PromptTemplate` (una plantilla de indicación) que tiene una casilla reservada para la pregunta.

```python
from langchain import PromptTemplate #Para definir plantillas de prompts

plantilla_colombiana = """Responde a la pregunta con base en el siguiente contexto, parce. Si la
pregunta no puede ser respondida con la información proporcionada, responde
con "No tengo ni idea, ome".

Contexto: {contexto}
Pregunta: {pregunta}

Respuesta (escribe como colombiano informal): """

#Definimos la plantilla de prompt
prompt_plantilla_colombiana = PromptTemplate(
    input_variables=["pregunta","contexto"],
    template=plantilla_colombiana
)
```

```python
from langchain import LLMChain

#Creamos una cadena de Question Answering
llm_gpt3_5_chain = LLMChain(
    prompt=prompt_plantilla_colombiana, #Plantilla de prompt
    llm=llm_gpt3_5 #LLM
)
```

```python
pregunta = "Qué son los LLMs?" #Pregunta
#pregunta="Cuantas sedes tiene la universidad Nacional de Colombia"
contexto="La Universidad Nacional de Colombia tiene 9 sedes a nivel nacional y una de ellas es Manizales,Caldas" #Contexto

answer=llm_gpt3_5_chain.run({"pregunta":pregunta,"contexto":contexto})
answer
```

```python
from langchain.chains.summarize import load_summarize_chain

#Creamos una cadena para la tarea de resumen de textos:
cadena_que_resume_documentos = load_summarize_chain(
    llm_gpt3_5, #LLM a utilizar
    chain_type="map_reduce" #Para que haga un resumen de cada página y luego haga un resumen de los respectivos resumenes generados
)
```

```python
summary=cadena_que_resume_documentos.run(data)
summary
```

Ahora, si necesito generar el resumen en español, puedo incluir mi propia plantilla de prompt donde le dé esta instrucción.

```python
from langchain import PromptTemplate #Para definir plantillas de prompts
```

```python
plantilla = """Escribe un resumen bien chido del siguiente rollo en español:

{text}

RESUMEN CORTO CON SLANG MEXICANO:"""

prompt = PromptTemplate(
    template=plantilla, #Plantilla del prompt
    input_variables=["text"] #Entradas a la plantilla del prompt
)
```

```python
#Cadena de resumen de textos:
cadena_que_resume_con_slang = load_summarize_chain(
    llm=llm_gpt3_5, #LLM a utilizar
    prompt=prompt,  #prompt
    verbose=True #Para imprimir el verbose
)
```

```python
summary=cadena_que_resume_con_slang.run(data) #Generamos el resumen
summary
```

```python
summary=cadena_que_resume_con_slang.run(data[:2]) #Generamos el resumen solo de las dos primeras páginas
summary
```
