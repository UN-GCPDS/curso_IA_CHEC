Streamlit es una herramienta de c√≥digo abierto dise√±ada para ayudar a los desarrolladores a crear aplicaciones web interactivas y visuales de manera muy r√°pida y sencilla, usando el lenguaje de programaci√≥n Python.

Streamlit te permite tomar ese c√≥digo y transformarlo en una aplicaci√≥n web funcional con solo unas pocas l√≠neas.

**¬øPara qu√© sirve Streamlit?**

- Crear interfaces visuales para proyectos en Python: Puedes mostrar gr√°ficos, botones, formularios, tablas, etc.
- Compartir proyectos interactivos: Los usuarios pueden interactuar con tus datos o modelos a trav√©s de la web sin saber programar.
- Prototipar r√°pido: En lugar de construir toda una p√°gina web desde cero, Streamlit facilita la creaci√≥n de prototipos de manera r√°pida.

## Creaci√≥n de un archivo `.py`

El archivo `.py` contendr√° todo el c√≥digo necesario para construir y ejecutar la aplicaci√≥n con **Streamlit**. En √©l, podr√°s incluir elementos como:

- Texto que deseas mostrar en la aplicaci√≥n.
- Comandos y l√≥gica para controlar el flujo de la app.
- Widgets interactivos (botones, sliders, etc.).
- Gr√°ficas, tablas y otras visualizaciones de datos.

Este archivo ser√° el coraz√≥n de la aplicaci√≥n, ya que define su estructura, contenido y comportamiento.

````python
%%writefile primer_streamlit.py
import streamlit as st

# T√≠tulo de la app
st.title("Mi primera aplicaci√≥n con Streamlit")

# Texto introductorio
st.write("¬°Bienvenidos a mi aplicaci√≥n interactiva!")

# Agregar un bot√≥n
if st.button("Presiona aqu√≠"):
    st.write("¬°El bot√≥n ha sido presionado!")

# Agregar un √°rea de entrada de texto
user_input = st.text_input("Ingresa tu nombre y presiona enter:", "")
if user_input:
    st.write(f"¬°Hola, {user_input}!")

# Agregar una secci√≥n para una calculadora simple
st.write("### Calculadora simple")

# Entradas para los n√∫meros
num1 = st.number_input("Ingresa el primer n√∫mero", value=0)
num2 = st.number_input("Ingresa el segundo n√∫mero", value=0)

# Bot√≥n para sumar los n√∫meros
if st.button("Sumar"):
    resultado = num1 + num2
    st.write(f"El resultado de la suma es: {resultado}")

# Texto final
st.write("Puedes incluir cualquier contenido en esta aplicaci√≥n, incluyendo c√°lculos, visualizaciones, e interacci√≥n con el usuario.")
````
